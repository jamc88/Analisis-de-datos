Desplegar un rango de columnas

```python
atletas.loc[["MEX"],"Genero":"Tiempo"]
```

Obtener la lista de mujeres que participaron en el marat√≥n

```python
atletas_Mujeres = atletas[atletas["Genero"]== "Female"]
atletas_Mujeres
```

Obtener una lista de los atletas mayores de 40 a√±os


```python
atletas_Edad = atletas[atletas["Edad"] > 40]
atletas_Edad
```

Buscamos participantes hombres √≥ participantes cuyo tiempo fu√© menor a 200 minutos

```python
atletas_hom_o_veloz = atletas[(atletas["Genero"] == "Male") | (atletas["Tiempo"] < 200)]

atletas_hom_o_veloz
```

Obtener una lista de las mujeres mayores de 40 a√±os

```python
atletas_mujeres_edad = atletas[(atletas["Genero"] == "Female") & (atletas["Edad"] > 40)]
atletas_mujeres_edad
```
