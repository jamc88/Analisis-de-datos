# TÃ©cnicas de Remplazo de valores perdidos

```python
import pandas as pd
import numpy as np 
```

```python
niveles=pd.read_csv("/Users/juanalbertomartinez/Desktop/RCH/remplazo_vp/Niveles de contaminacion.csv", engine="python")
```

```python
niveles.info()
```

```python
niveles
```

Remplazar los valores perdidos con un valor fijo

```python
niveles1=niveles
niveles1.fillna(0) #remplaza los valores faltantes con un "0"
```

Reemplazar los valores perdidos con la media de cada columna

```python
niveles2=niveles.apply(lambda x: x.fillna(x.mean()), axis=0)
niveles2
```

Reemplazar los valores perdidos con la mediana de cada columna


```python
niveles3=niveles.apply(lambda x: x.fillna(x.median()), axis=0)
niveles3
```

Remplazando los valores perdidos de la columna ProfesiÃ³n con la palabra "Desconocida"
