# Categorización de Variables

```python
import pandas as pd
import math
print("Pandas Versión", pd.__version__)
```

```python
atletas = pd.read_csv("/Users/juanalbertomartinez/Desktop/RCH/Categorizacion_variables/Corredores.csv",
                     index_col = 0) # Con index_col=0 le indicamos que las filas tienen un nombre

atletas
```

```python
atletas.info()
```

Definimos los limites que tendrá cada categoria de la variable Tiempo

```python
valores = [-math.inf, 200, 300, math.inf]
```

Indicamos cuales serán los nomres de las categorias 

```python
categorias = ["Veloz", "Regular", "Lento"]
```

Agregamos una columna nueva en la que estarán indicadas las categorias asignadas

```python
atletas["Velocidad"] = pd.cut(x = atletas["Tiempo"], bins = valores, labels = categorias)

atletas
```
